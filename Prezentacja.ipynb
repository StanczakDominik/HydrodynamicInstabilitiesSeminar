# What is hydrodynamics?

* language - PDEs, 3+1D or less - symmetry

## Navier-Stokes equations 
    
$$ \rho \frac{\partial{\vec{u}}}{\partial t} +  \rho (\vec{u} \cdot \nabla) \vec{u} - \rho \nu \nabla^2 \vec{u} = - \nabla p + \rho \vec{g}$$

* Momentum conservation equation, in essence

\begin{align}
 \underbrace{\rho \frac{\partial \vec{u}}{\partial t}}_{
 \begin{smallmatrix}
 \text{Variation}
 \end{smallmatrix}} +
 \underbrace{\rho (\vec{u} \cdot \nabla) \vec{u}}_{
 \begin{smallmatrix}
 \text{Convection}
 \end{smallmatrix}}
 -\underbrace{\rho \nu \, \nabla^2 \vec{u}}_{\text{Diffusion}}=
 \underbrace{-\nabla p}_{
 \begin{smallmatrix}
 \text{Internal} \\
 \text{source}
 \end{smallmatrix}} +
 \underbrace{\rho \vec{g}}_{
 \begin{smallmatrix}
 \text{External} \\
 \text{source}
 \end{smallmatrix}}
\end{align}

* $\vec{u}(\vec{r}, t)$ - the fluid velocity
* $\rho(\vec{r}, t)$ - the density of the fluid
* $\nu$ - fluid viscosity
* $p(\vec{r}, t)$ - internal potentials (pressure!)
* $\vec{g}(\vec{r})$ - external forces (gravity)
* In full form the diffusion and pressure terms are $\nabla \cdot \hat{\tau}$, a tensor expression

### Advection and convection

\begin{align}
 \underbrace{\frac{\partial \vec{u}}{\partial t}}_{
 \begin{smallmatrix}
 \text{Variation}
 \end{smallmatrix}} +
 \underbrace{(\vec{u} \cdot \nabla) \vec{u}}_{
 \begin{smallmatrix}
 \text{Convection}
 \end{smallmatrix}} = 0
\end{align}

\begin{align}
 \underbrace{\frac{\partial \rho}{\partial t}}_{
 \begin{smallmatrix}
 \text{Variation}
 \end{smallmatrix}} +
 \underbrace{(\vec{u} \cdot \nabla) \rho}_{
 \begin{smallmatrix}
 \text{Advection}
 \end{smallmatrix}} = 0
\end{align}

Nonlinear equation!

* Convection - movement of fluid itself
* Advection - movement of material dissolved in fluid (paint)

### Diffusion

\begin{align}
 \underbrace{\frac{\partial \vec{u}}{\partial t}}_{
 \begin{smallmatrix}
 \text{Variation}
 \end{smallmatrix}}
 -\underbrace{\nu \, \nabla^2 \vec{u}}_{\text{Diffusion}}= 0
\end{align}

* Laplacian of a vector quantity
* Returns "three scalar laplacians"

## Continuity of mass equation

$$\frac{\partial \rho}{\partial t} + \nabla \cdot (\rho \vec{u}) = 0$$

With the constant density assumption:

$$\rho = \text{const} \implies \nabla \cdot (\rho \vec{u}) = 0 \implies \nabla \cdot \vec{u} = 0$$

## Instability (generic term)

> Some parameters of a system growing without bounds!

Loose definition!

## Fluid stability

Practical definition from deterministic chaos:

* two slightly different initial conditions converge to the same solution $\implies$ **stability**
* two slightly different initial conditions converge to **different** solutions $\implies$ **instability**

## Linear Stability Analysis

1. Assume some base state and small perturbation around it

2. For the perturbation assume Fourier expansion $\exp\left(i (\vec{k} \cdot \vec{x} - \omega t) \right)$
    * $\vec{k}$ - perturbation wavenumber
    * $\omega$ - angular (time) frequency

3. See if the perturbation grows, dies or stays constant

### Stability analysis for Rayleigh-Taylor

For each layer we take
$$\vec{u} = (0 + u(\vec{r}, t), 0, 0 + w(\vec{r}, t))$$

(2D situation)

Plug this into $\nabla \cdot \vec{u} = 0$:

$$\frac{\partial u}{\partial x} + \frac{\partial w}{\partial z} = 0$$

## Navier-Stokes (Euler) Equations for Rayleigh Taylor - written explicitly

* remove viscosity $- \nu \nabla^2 \vec{u}^2$ (thus 2D, inviscid, incompressible)

$$\rho \frac{\partial{\vec{u}}}{\partial t} + \rho (\vec{u} \cdot \nabla) \vec{u} = - \nabla p + \rho \vec{g}$$


Assume stratified layers ("Boussinesq approximation") $\rho = \rho(z) + \rho'(x, z, t)$, linearize $\rho' u$

\begin{align}
\rho(z) \frac{\partial{u}}{\partial t} + \rho(z)(u \partial_x + w \partial_z) u &= - \partial_x p + 0 \\
\rho(z) \frac{\partial{w}}{\partial t} + \rho(z)(u \partial_x + w \partial_z) w &= - \partial_z p - (\rho(z) + \rho')g
\end{align}

Assume hydrostatic pressure $p = p(z) + p'(x, z, t), \frac{dp}{dz} = - \rho g$

\begin{align}
\rho(z) \frac{\partial{u}}{\partial t} + \rho(z)(u \partial_x + w \partial_z) u &= - \partial_x p'\\
\rho(z) \frac{\partial{w}}{\partial t} + \rho(z)(u \partial_x + w \partial_z) w &= - \partial_z p' - \rho'g
\end{align}

Linearize - neglect terms like $u \partial_x u$ (second order)

\begin{align}
\rho(z)\frac{\partial{u}}{\partial t} &= - \partial_x p'\\
\rho(z)\frac{\partial{w}}{\partial t }&= - \partial_z p' - \rho'(x, z, t)g \\
\frac{\partial u}{\partial x} + \frac{\partial w}{\partial z} &= 0
\end{align}

As the fourth equation, we'll take the mass continuity equation for the density with $\rho'$ perturbation

$$\frac{\partial (\rho(z) + \rho')}{\partial t} + \nabla \cdot ((\rho(z) + \rho') \vec{u}) = 0$$
$$\frac{\partial (\rho(z) + \rho')}{\partial t} + \partial_x  ((\rho(z) + \rho') u) + \partial_z ((\rho(z) + \rho') w)= 0$$
$$\frac{\partial \rho'}{\partial t} + \rho(z) (\partial_x u + \partial_w z) + w \partial_z \rho = 0$$
$$\frac{\partial \rho'}{\partial t} +  w \frac{d\rho}{dz} = 0$$

Thus:

\begin{align}
\rho(z)\frac{\partial{u}}{\partial t} &= - \partial_x p'\\
\rho(z)\frac{\partial{w}}{\partial t }&= - \partial_z p' - \rho'(x, z, t)g \\
\frac{\partial u}{\partial x} + \frac{\partial w}{\partial z} &= 0 \\
\frac{\partial \rho'}{\partial t} &= - w \frac{d\rho}{dz}
\end{align}

### Fourier expansion

$$u(x, z, t) = u_k(z) \exp(ikx + n t)$$ 
$$w(x, z, t) = w_k(z) \exp(ikx + n t)$$ 
$$p'(x, z, t) = p'_k(z) \exp(ikx + n t)$$ 
$$\rho(x, z, t) = \rho'_k(z) \exp(ikx + n t)$$ 

We expect instability, thus $nt$ instead of $-i \omega t$

\begin{align}
ik u + \frac{\partial w}{\partial z} &= 0 \\
\rho n u &= - i k p\\
\rho n w &= - \rho' g  - \partial_z p' \\
n \rho' &= -w \frac{d\rho}{dz}
\end{align}

This is a set of algebraic equations that simplifies to

$$\frac{d}{dz}(\rho \frac{dw}{dz}) - \rho k^2 w = - w g \frac{k^2}{n^2} \frac{d \rho}{dz}$$ 

* fluid 2 on top, fluid 1 on the bottom
* directly above and below boundary, density is constant, we get

$$\frac{d^2 w}{dz^2} - k^2 w = 0$$
$$w_2 = w_0 \exp(-kz) (z>0)$$
$$w_1 = w_0 \exp(kz) (z<0)$$

Integrate previous equation across the interface between layers $dz = \pm \epsilon$

$$\int d (\rho \frac{dw}{dz}) - \int \rho k^2 w dz = - \int w g \frac{k^2}{n^2} d \rho$$

Second term is zero,

$$\rho_2 (-kw) - \rho_1 (kw) = -wg \frac{k^2}{\omega^2} (\rho_2 - \rho_1)$$

$kw$ simplifies:

$$-(\rho_2 + \rho_1) = -g \frac{k^2}{n^2} (\rho_2 - \rho_1)$$

## The result of linear stability analysis

Solving for $n$:

$$n = \sqrt{gk \frac{\rho_2 - \rho_1}{\rho_2 + \rho_1}} = C_+ \sqrt{k (\rho_2 - \rho_1)}$$

* Positive (perturbation grows) for any non-zero wavevector $k$ and $\rho_2 > \rho1$ unstable
* Imaginary (perturbation oscillates) for $\rho_2 < \rho_1$ - stable
* Zero for equal densities or no perturbation - no interaction whatsoever

... but that's all the classical analysis can tell us!
